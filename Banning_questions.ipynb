## If Teemo is banned, is his team more or less likely to win (or neither)?

First, we need to extract matches in which Teemo is banned. Let's keep blue and red separate for now.

```python
blueban = dtr.Bundle([matches.categories.groupby(ban)['Teemo'] for ban in ('Blue Ban 1', 'Blue Ban 2', 'Blue Ban 3')])
```

```python
blueban
```

```python
redban = dtr.Bundle([matches.categories.groupby(ban)['Teemo'] for ban in ('Red Ban 1', 'Red Ban 2', 'Red Ban 3')])
```

```python
redban
```

If we wanted all the matches where Teemo was banned by at least one team, then we could get it with:

```python
redban + blueban
```

What if we wanted all matches where Teemo was banned by *both* teams:

```python
redban & blueban
```

This is expected, because both teams can't ban same champion. :)

For red team bans, how many of those matches did the red team lose?

```python
import pandas as pd
```

```python
df = pd.read_csv(redban[0]['match.csv'].abspath)
df
```

```python
def blue_win(df):
    return 'blue' in df[df['Win'] == True]['Side'].iloc[0]
```

```python
redban_blue_wins = [blue_win(pd.read_csv(match['match.csv'].abspath)) for match in redban]
redban_blue_wins
```

So now we have the matches in which red team banned Teemo and either won or lost. What now?

Did red win more often than they lost? Here are the matches red won:

```python
len(redban - redban[redban_blue_wins])
```

Here are the number of matches red lost:

```python
len(redban[redban_blue_wins])
```

# Who is the most banned character? Does banning this character make your team more likely to win?

We want to take to task some conventional wisdom that may be floating around the community about which characters you should ban if you want to increase your chances to win. Let's see if this holds water.

```python
len(matches)
```

First we need to find the most banned character.

```python
counts = pd.DataFrame({ban: {key: len(value) for key, value in matches.categories.groupby(ban).items()} for ban in ('Blue Ban 1', 'Blue Ban 2', 'Blue Ban 3', 'Red Ban 1', 'Red Ban 2', 'Red Ban 3')})
counts
```

```python
counts.sum(axis=1).sort_values(ascending=False)
```

So we have our answer. Most banned is **Rengar**.

# In a game where Rengar wasn't banned, was he picked? Was the team more likely win?

First we need matches where Rengar was not banned.

```python
no_banned_rengar = matches - dtr.Bundle([matches.categories.groupby(ban)['Rengar'] for ban in ('Blue Ban 1', 'Blue Ban 2', 'Blue Ban 3', 'Red Ban 1', 'Red Ban 2', 'Red Ban 3')])
no_banned_rengar
```

So now, which matches was Rengar present in?

```python
df = pd.read_csv(no_banned_rengar[0]['match.csv'].abspath)
df
```

```python
len(df[df['Champion'] == 'Rengar']) == 1
```

```python
def rengar_present(df):
    return len(df[df['Champion'] == 'Rengar']) == 1
```

```python
has_rengar = [rengar_present(pd.read_csv(match['match.csv'].abspath)) for match in no_banned_rengar]
has_rengar
```

```python
matches_rengar = no_banned_rengar[has_rengar]
```

Here are the matches where Rengar was used:

```python
len(matches_rengar)
```

## If a team had a Rengar, were they more likely to win?

```python
matches_rengar
```

```python
def rengar_won(df):
    return bool(df[df['Champion'] == 'Rengar']['Win'].iloc[0])
```

```python
won_with_rengar = [rengar_won(pd.read_csv(match['match.csv'].abspath)) for match in matches_rengar]
won_with_rengar
```

Number of wins:

```python
sum(won_with_rengar)
```

```python
len(matches_rengar[won_with_rengar])
```

Number of losses:

```python
len(won_with_rengar) - sum(won_with_rengar)
```

```python
len(matches_rengar - matches_rengar[won_with_rengar])
```

Doesn't really look like he confers too much of an advantage, but perhaps a slight one?
